# Motor Vehicle Collisons Data by NYPD

NYC OpenData: https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95/data

The Motor Vehicle Collisions crash table contains details on the crash event. Each row represents a crash event. The Motor Vehicle Collisions data tables contain information from all police reported motor vehicle collisions in NYC.

## Column(s) Descriptions:

    CRASH DATE: The date of the collision.
    CRASH TIME: The time of the collision.
    BOROUGH: The city borough in which the collision occured.
    LATITUDE, LONGITUDE and LOCATION: Geographical coordinates of the collision.
    ON STREET NAME: Street on which the collision occurred.
    CROSS STREET NAME: Nearest cross street to the collision.
    OFF STREET NAME: Street address (if known).
    NUMBER OF PERSONS INJURED, NUMBER OF PERSONS KILLED, NUMBER OF PEDESTRIANS INJURED, NUMBER OF PEDESTRIANS KILLED, NUMBER OF CYCLIST INJURED, NUMBER OF CYCLIST KILLED, NUMBER OF MOTORIST INJURED and NUMBER OF MOTORIST KILLED: Details about the number of people injured or killed in the accident.
    CONTRIBUTING FACTOR VEHICLE 1-5: Factors contributing to the collision for designated vehicle.
    COLLISION_ID: Unique record code generated by the system.
    VEHICLE TYPE CODE 1-5: Type of vehicle based on the selected vehicle category

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
Motor_Crash = r'C:\Users\jadaj\OneDrive\Desktop\SCHOOL\Motor_Vehicle_Collisions_-_Crashes.csv'
df = pd.read_csv(Motor_Crash)
df
```

## Pre Processing & Data Cleaning

```python
df.info()
```

```python
df.isnull().sum().plot.bar()
```

```python
df.columns
```

```python
drop_null = df.isnull().sum() / len(df) < 0.33
df = df.loc[:, drop_null]

df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])
df['CRASH TIME'] = pd.to_datetime(df['CRASH TIME'])
df['YEAR'] = df['CRASH DATE'].dt.year
df['MONTH'] = df['CRASH DATE'].dt.month
df['HOUR'] = df['CRASH TIME'].dt.hour
month_dict = {1:'Jan',2:'Feb',3:'Mar',4:'April',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'}
df['MONTH'].replace(month_dict, inplace = True)


drop_cols = ['CRASH DATE', 'CRASH TIME', 'ZIP CODE', 'LOCATION', 'CONTRIBUTING FACTOR VEHICLE 2', 'VEHICLE TYPE CODE 2']
df.drop(drop_cols, axis = 1, inplace = True)

new_cols = ['YEAR', 'MONTH', 'HOUR', 'BOROUGH', 'LATITUDE', 'LONGITUDE', 'ON STREET NAME',
       'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
       'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
       'CONTRIBUTING FACTOR VEHICLE 1', 'COLLISION_ID', 'VEHICLE TYPE CODE 1']

crash_data = df.reindex(columns = new_cols)
crash_data
```

## Accident Analysis

```python
borough_accidents = crash_data.groupby(['BOROUGH']).size().reset_index(name='NoOfAccidents')
year_accidents = crash_data.groupby(['YEAR']).size().reset_index(name='NoOfAccidents')
byhour_accidents = crash_data.groupby(['HOUR']).size().reset_index(name='NoOfAccidents')
```

```python
plt.figure(figsize=(24,12))
ax1=plt.subplot2grid((2,2),(0,0))
sns.barplot(x= 'BOROUGH',y='NoOfAccidents',data= borough_accidents).set(title='Number of Accidents In Each Borough')

plt.figure(figsize=(24,12))
ax2=plt.subplot2grid((2,2),(0,1))
sns.barplot(x= 'YEAR',y='NoOfAccidents',data= year_accidents).set(title='Number of Accidents Per Year')

plt.figure(figsize=(24,12))
ax3=plt.subplot2grid((2,2),(0,0))
sns.barplot(x= 'HOUR',y='NoOfAccidents',data= byhour_accidents).set(title='Number of Accidents Per Hour')
```

### Injuries and Fatalities

```python
injuries_and_fatalities = crash_data.groupby(['BOROUGH', 'YEAR'])['NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED'].sum().reset_index()
injuries_and_fatalities['Total Accidents'] = crash_data.groupby(['BOROUGH', 'YEAR']).size().reset_index(name='NoOfAccidents')['NoOfAccidents']
injuries_and_fatalities
```

```python
injuries_and_fatalities['Injury%'] = round((injuries_and_fatalities['NUMBER OF PERSONS INJURED'] / 
                                            injuries_and_fatalities['Total Accidents'] * 100), 0)
injuries_and_fatalities['Fatality%'] = round((injuries_and_fatalities['NUMBER OF PERSONS KILLED'] / 
                                              injuries_and_fatalities['Total Accidents'] * 100), 2)

injuries_and_fatalities
```

```python
plt.figure(figsize=(24, 12))
sns.set(style='whitegrid', font_scale=2)
sns.lineplot(y='NUMBER OF PERSONS INJURED', x='YEAR', hue='BOROUGH', data= injuries_and_fatalities)

plt.figure(figsize=(24, 12))
sns.set(style='whitegrid', font_scale=2)
sns.lineplot(y='NUMBER OF PERSONS KILLED', x='YEAR', hue='BOROUGH', data= injuries_and_fatalities)

plt.figure(figsize=(24, 12))
sns.set(style='whitegrid', font_scale=2)
sns.lineplot(y='Injury%', x='YEAR', hue='BOROUGH', data= injuries_and_fatalities)

plt.figure(figsize=(24, 12))
sns.set(style='whitegrid', font_scale=2)
sns.lineplot(y='Fatality%', x='YEAR', hue='BOROUGH', data= injuries_and_fatalities)
```
